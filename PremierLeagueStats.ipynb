```python
#Comparing wins from 2005-06 and 2021-21 
season_2005_2006 = df[df['Season'] == '2005-06']
total_wins_2005_2006 = season_2005_2006['W'].sum()

# Calculating the total number of teams for the 2005-2006 season
total_teams_2005_2006 = season_2005_2006.shape[0]
# Calculating the mean number of wins for the 2005-2006 season
mean_wins_2005_2006 = total_wins_2005_2006 / total_teams_2005_2006
print("Mean number of wins for the 2005-2006 season:", mean_wins_2005_2006)

season_2020_2021 = df[df['Season'] == '2020-21']
# Calculating the total number of wins for the 2020-2021 season
total_wins_2020_2021 = season_2020_2021['W'].sum()
# Calculating the total number of teams for the 2020-2021 season
total_teams_2020_2021 = season_2020_2021.shape[0]

# Calculating the mean number of wins for the 2020-2021 season
mean_wins_2020_2021 = total_wins_2020_2021 / total_teams_2020_2021
print("Mean number of wins for the 2020-2021 season:", mean_wins_2020_2021)

# Calculating the difference in mean number of wins between the two seasons
mean_difference = mean_wins_2020_2021 - mean_wins_2005_2006
print("Mean difference between the 2005-2006 season and the 2020-2021 season:", mean_difference)
# Check if the mean difference is negative and print the corresponding message
if mean_difference < 0:
    print("It indicates that in the 2005 season, teams had won more games than the recent season.")

```

```python
#Comparing Pts from 2005-06 and 2021-21 
season_2005_2006 = df[df['Season'] == '2005-06']

total_points_2005_2006 = season_2005_2006['Pts'].sum()

total_teams_2005_2006 = season_2005_2006.shape[0]

mean_points_2005_2006 = total_points_2005_2006 / total_teams_2005_2006

print("Mean number of points for the 2005-2006 season:", mean_points_2005_2006)

season_2020_2021 = df[df['Season'] == '2020-21']

total_points_2020_2021 = season_2020_2021['Pts'].sum()

total_teams_2020_2021 = season_2020_2021.shape[0]

mean_points_2020_2021 = total_points_2020_2021 / total_teams_2020_2021

print("Mean number of points for the 2020-2021 season:", mean_points_2020_2021)

mean_difference_points = mean_points_2020_2021 - mean_points_2005_2006

print("Mean difference between the 2005-2006 season and the 2020-2021 season:", mean_difference_points)

if mean_difference_points < 0:
    print("It indicates that the 2005 season, teams scored more points.")
```

```python
#Count how many times the "Big 6" were actually in the top 6 from 2000-2022
teams_to_count = ['Manchester United', 'Liverpool', 'Arsenal', 'Tottenham Hotspur', 'Manchester City', 'Chelsea']
team_counts = {}

for team in teams_to_count:
    #Count how many times the BIG 6 teams in the array, are in the top 6 and count
    team_df = df[(df['Team'] == team) & (df['Pos'] <= 6)]
    count = len(team_df)
    team_counts[team] = count

for team, count in team_counts.items():
    print(f"{team}: {count} times in the top 6 position")
```

```python
#How many premier league titles have the "Big 6" won from 2000-2022
big_six_teams = ['Manchester United', 'Liverpool', 'Arsenal', 'Tottenham Hotspur', 'Manchester City', 'Chelsea']

team_wins_count = {}

for team in big_six_teams:
     #Count how many times the BIG 6 teams in the array, are in the first position and count
    team_wins = df[(df['Team'] == team) & (df['Pos'] == 1)]
    team_win_count = len(team_wins)
    team_wins_count[team] = team_win_count

for team, count in team_wins_count.items():
    print(f"{team} won the league {count} times")
```

```python
#Since 21 premier leagues were won by the "Big 6", let's take a look at which team outside the big 6 won the premier league
#selects the teams outside of the "big six" that finished in the first position in the league.
non_big_six_teams = df[~df['Team'].isin(big_six_teams)]
non_big_six_wins = non_big_six_teams[non_big_six_teams['Pos'] == 1]

if len(non_big_six_wins) > 0:
    print(f"The team(s) outside the Big 6 that won the Premier League: {', '.join(non_big_six_wins['Team'].unique())}")
else:
    print("No team outside the Big 6 has won the Premier League.")
```

```python
#Standard deviation of Pts, combined top 6
mean_pts = combined_table['Pts'].mean()
squared_diffs = (combined_table['Pts'] - mean_pts) ** 2
mean_squared_diffs = squared_diffs.mean()

std_dev = mean_squared_diffs ** 0.5
print("Standard Deviation:", std_dev)
```

```python
#Sample variance of points, in combined top 6 table
mean_pts = combined_table['Pts'].mean()

# Calculate the squared differences between each data point and the mean
squared_diffs = (combined_table['Pts'] - mean_pts) ** 2
sum_squared_diffs = squared_diffs.sum()

# Calculate the sample variance (divide the sum of squared differences by n - 1)
n = len(combined_table['Pts'])
sample_variance = sum_squared_diffs / (n - 1)

print("Sample Variance:", sample_variance)

```

```python
#Coefficient of variance of the Points
mean_pts = combined_table['Pts'].mean()
cv = (std_dev / mean_pts) * 100

print("Coefficient of Variation:", cv, "%")
```

```python
#Goal difference between Man united (2000-01 winner) and Man City (2021-22 winner)
gd_2000_01_manchester_united = df_2000_01[df_2000_01['Team'] == 'Manchester United']['GD'].values[0]
gd_2021_22_manchester_city = df_2021_22[df_2021_22['Team'] == 'Manchester City']['GD'].values[0]

# Compare the GD for Manchester United in 2000-2001 season and Manchester City in 2021-2022 season
print("Goal Difference for Manchester United in the 2000-2001 season:", gd_2000_01_manchester_united)
print("Goal Difference for Manchester City in the 2021-2022 season:", gd_2021_22_manchester_city)
gd_difference = gd_2000_01_manchester_united - gd_2021_22_manchester_city
print("Difference in Goal Difference (Manchester United - Manchester City):", gd_difference)
#This indicates that Manchester City had a much better goal ratio than man united
```

```python
# Sum up the points for the 2000-2001 season
total_pts_2000_01 = df_2000_01['Pts'].sum()
print("Total points for the 2000-2001 season:", total_pts_2000_01)
# Sum up the points for the 2021-2022 season
total_pts_2021_22 = df_2021_22['Pts'].sum()
print("Total points for the 2021-2022 season:", total_pts_2021_22)
```

```python
# Calculate the covariance between the points for the two seasons
covariance = df_2000_01['Pts'].cov(df_2021_22['Pts'])
print("Covariance between the 'Pts' for the 2000-2001 and 2021-2022 seasons:", covariance)
```

```python
# Calculate the correlation coefficient between the points for the two seasons
correlation_coefficient = df_2000_01['Pts'].corr(df_2021_22['Pts'])
print("Correlation coefficient between the 'Pts' for the 2000-2001 and 2021-2022 seasons:", correlation_coefficient)
```
